## 바 그래프

```python
data = {
    '바나나' : 3, '사과' :2, '오랜지' : 10
}
# 선 : plot, 바 : bar
print(data.keys(),data.values())
#plt.bar(data.keys(), data.values())
plt.bar(data.keys(), data.values())
plt.title('과일 개수')
plt.xticks([0,1,2], data.keys())
plt.yticks(sorted(data.values()))
plt.grid(True) # 격자 표시
plt.xlabel('과일')
plt.ylabel('개수')
plt.show()
```

## 가로로 그리기

```python
data = {
    '바나나' : 3, '사과' :2, '오랜지' : 10
}
# 선 : plot, 세로바 : bar, 가로 바:barh
print([0,1,2],data.values())
#plt.bar(data.keys(), data.values())
plt.barh([0,1,2],data.values())   #barh로 해야 가로로 된다.
plt.title('과일 개수')
plt.yticks([0,1,2], data.keys())
plt.xticks(sorted(data.values()))
plt.grid(True) # 격자 표시
plt.xlabel('과일')
plt.ylabel('개수')
plt.show()
```

# 파이차트

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grape']
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['silver', 'gold', 'whitesmoke', 'lightgray']
#파이차트 매서드 plt.pie
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
```

```python
 wedgeprops={'width': 0.7, 'linewidth': 5}

plt.pie(ratio, labels=labels, 
        autopct ='%.1f%%',
        startangle=0,
        counterclock=True,
        colors=colors,
        wedgeprops=wedgeprops
)
```

## 박스 그래프

```python
np.random.seed(0)
plt.style.use('default')

data_a = np.random.normal(0,2.0,1000)
data_b = np.random.normal(-3.0,1.5,500)
data_c = np.random.normal(1.2,1.5,1500)

fig, ax = plt.subplots()
# 박스 그래프 : boxplot
ax.boxplot([data_a, data_b, data_c], notch=True, whis=1.5, vert = True) #notch는 신뢰 구간, vert=false는 가로
ax.set_ylim(-10.0, 10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

## 바이올린 플롯(Violin Plot)

```python
plt.style.use('default')
# 데이터를 준비
np.random.seed(0)
data_a = np.random.normal(0,2.0,1000)
data_b = np.random.normal(-3.0,1.5,500)
data_c = np.random.normal(1.2,1.5,1500)

violin = plt.violinplot([data_a, data_b, data_c], showmeans = True, showextrema = True, 
                        showmedians = True)
plt.xticks(np.arange(1,4), labels=['A','B','C'])
plt.show()
```

```python
plt.style.use('default')
# 데이터를 준비
np.random.seed(0)
data_a = np.random.normal(0,2.0,1000)
data_b = np.random.normal(-3.0,1.5,500)
data_c = np.random.normal(1.2,1.5,1500)

violin = plt.violinplot([data_a, data_b, data_c], showmeans = True, showextrema = True, 
                        showmedians = True)
plt.xticks(np.arange(1,4), labels=['A','B','C'])
violin['bodies'][0].set_facecolor('red')
violin['cbars'].set_edgecolor('red')
violin['cmaxes'].set_edgecolor('orange')
violin['cmins'].set_edgecolor('black')
violin['cmeans'].set_edgecolor('red')
violin['cmedians'].set_edgecolor('black')
plt.show()
```
